```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=1000, centers=3, n_features=2)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1], c=y, cmap='Accent')
```

```python
# Standardization-- feature scaling technique
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
X_train_scalar = scalar.fit_transform(X_train)
X_test_scalar = scalar.transform(X_test)
```

```python
X_train_scalar
```

```python
X_test_scalar
```

```python
# Importing KMeans from sklearn.cluster

from sklearn.cluster import KMeans
```

```python
# Elbow Method to select the K value
# wcss - within clusters sum of square

wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X_train_scalar)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# plot elbow curve

sns.lineplot(wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
```

```python
kmeans.fit_predict(X_train_scalar)
```

```python
# predict X_test_scalar
y_pred = kmeans.predict(X_test_scalar)
```

```python
y_pred
```

```python
sns.scatterplot(x = X_test[:,0], y = X_test[:,1], c = y_pred, cmap='Accent')
plt.show()
```

```python
# Validating the k value
# 1- kneelocator
# 2- Silhouette scoring
```

```python
# 1 - kneelocator
from kneed import KneeLocator

kl= KneeLocator(range(1,11), wcss, curve='convex', direction='decreasing')
```

```python
print(kl.elbow)
```

```python
# 2 - silhouette scoring

from sklearn.metrics import silhouette_score

silhouette_coefficient = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X_train_scalar)
    score = silhouette_score(X_train_scalar, kmeans.labels_)
    silhouette_coefficient.append(score)
```

```python
silhouette_coefficient
```

```python
sns.lineplot(silhouette_coefficient)
plt.show()
```
